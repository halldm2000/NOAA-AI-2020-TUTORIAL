### **Using Python: Plot a simple curve**

- Let's warm up with **Python** by doing something simple, like plotting a curve.


```
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10,10,100)
y = x**2

plt.plot(x,y);
```

- **`Numpy`** is a numerical python library for working with arrays and mathematical expressions
- **`MatPlotLib`** is a plotting library that reproduces many of the plotting functions of Matlab 


### **Pactice: Improve the plot**
- Let's practice a bit by trying to make a nicer plot.


```
import numpy as np
import matplotlib.pyplot as plt

# define the data
x  = np.linspace(-10,10,1000)
y1 = np.sin(3*x)*np.exp(-x**2/(5)**2)
y2 = 0.6*y1

# plot the curves
fig = plt.figure(figsize=(12,6))
plt.fill_between(x,y1,y2,color='k',alpha=0.1)
plt.plot(x,y1,'-', linewidth=1, label='y1')
plt.plot(x,y2,'-', linewidth=1, color='gray',label='y2')

# add some decorators
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True,alpha=0.3)
plt.legend()
plt.title('Some Kind of Wave Packet',fontsize=18);
```
